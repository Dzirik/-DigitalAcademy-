## Třída

> **POZOR** Upravujte pouze kód mezi:

> #Úloha n - Vykreslení dat ##############################
> #zde vložte kód
> ########################################################

```python
class kStredu:
    #data -----------------------------------------------------------------------------------------
    X = []  #proměnná pro uložení dat jako numpy array
    stred = [] #proměnná pro uložení středů shluků, v průběhu se mění
    prislusnost = [] #proměnná pro uložení příslušnosti pozorování ke shlukům
    hodnotaUceloveFunkce = float("Inf") #hodnota účelové funkce
    k = 0 #počet shluků
    pIteraci = 0 #počet iterací
    
    #při inicializace vložím uložím do proměnné k počet shluků ------------------------------------
    def __init__(self,k):
        #Úloha 1 - Předání dat #################################
        #zde vložte kód
        
        ########################################################
     
    #vložení dat ----------------------------------------------------------------------------------
    def vlozData(self,X):
        #Úloha 1 - Předání dat #################################
        #zde vložte kód
        
        ########################################################
    
        self.prislusnot = X.shape[0]*[0]
    
    #inicializace středů --------------------------------------------------------------------------
    def inicializujStredy(self,cislaPozorovani):
        self.stred=self.X[cislaPozorovani,:]
        
    #přiřazení - E-step ---------------------------------------------------------------------------
    def prirazeni(self):
        #inicializace pole pro uložení vzdáleností
        vzdalenost = np.array([[0.0 for col in range(self.k)] for row in range(self.X.shape[0])])
        
        #Úloha 4 - přiřazení ###################################
        #zde vložte kód
        #procházím shluky
        
        ######################################################## 
        
        #uložím si příslušnost ke shlukům
        self.prislusnost = vzdalenost.argmin(axis=1)
        
        #přidám hodnotu účelové funkce
        self.hodnotaUceloveFunkce = sum(vzdalenost.min(axis=1))
    
    #optimalizace - M-step ------------------------------------------------------------------------
    def optimalizace(self):
        #Úloha 5 - optimalizace ###############################
        
        #######################################################
        
        #inkrementace počtu iterací
        self.pIteraci+=1
            
    #vykreslení dat -------------------------------------------------------------------------------
    def vykresliData(self):
        #Úloha 2 - Vykreslení dat ##############################
        #zde vložte kód
        
        ########################################################
        
        #vynucené vykreslení
        plt.show()
        
    #vykreslení shluků ----------------------------------------------------------------------------
    def vykresliShlukovani(self):
        #Úloha 3 - Vykreslení shluků ###########################
        #zde vložte kód
        
        ########################################################
        
        #vynucené vykreslení
        plt.show()
        
    #hodnota účelové funkce ----------------------------------------------------------------------
    def ucelovaFunkce(self):
        print(self.hodnotaUceloveFunkce)
```
