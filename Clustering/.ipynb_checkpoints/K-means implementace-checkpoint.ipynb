```python
import matplotlib.pyplot as plt
import numpy as np

class kStredu:
    X = []
    stred = []
    prislusnost = []
    vzdalenost = []
    hodnotaUceloveFunkce = float("Inf")
    k = 0
    pIteraci = 0
    
    #při inicializace vložím uložím do proměnné k počet shluků
    def __init__(self,k):
        self.k = k
     
    #vložení dat
    def vlozData(self,X):
        self.X = X
        self.prislusnot = X.shape[0]*[0]
        self.vzdalenost = np.array([[0.0 for col in range(self.k)] for row in range(X.shape[0])])
        
    #vykreslení dat
    def vykresliData(self):
        plt.scatter(X[:,0],X[:,1])
        plt.show()
    
    #inicializace středů
    def inicializujStredy(self,cislaPozorovani):
        self.stred=self.X[cislaPozorovani,:]
        
    #přiřazení - E-step
    def prirazeni(self):
        #jsem v další iteraci
        for i in range(0,self.k):
            rozdil = self.X-np.array([self.stred[i,:]]*self.X.shape[0])
            #print(rozdil)
            #print(np.array([np.linalg.norm(x) for x in rozdil]))
            #self.X[:,i] = list([np.linalg.norm(x) for x in rozdil])
            self.vzdalenost[:,i] = list([np.linalg.norm(x) for x in rozdil])  
        #print(self.vzdalenost.argmin(axis=1))
        self.prislusnost = self.vzdalenost.argmin(axis=1)
        self.hodnotaUceloveFunkce = sum(self.vzdalenost.min(axis=1))
        #print(self.vzdalenost.min(axis=1))
        #print(sum(self.vzdalenost[:,self.prislusnost]))
        #print(self.vzdalenost)
    
    #optimalizace - M-step
    def optimalizace(self):
        for i in range(0,self.k):
            self.stred[i,:]=self.X[self.prislusnost == i,:].mean(axis=0)
        
    #vykreslení shluků
    def vykresliShlukovani(self):
        plt.scatter(self.X[:,0],self.X[:,1],c=self.prislusnost,cmap='rainbow')
        plt.scatter(self.stred[:, 0], self.stred[:, 1], s = 300, c = "black",edgecolors="black",marker="+")
        plt.show()
        '''
        plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
        plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
        plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
        plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
        plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
        plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
        plt.title('Clusters of customers')
        plt.xlabel('Annual Income (k$)')
        plt.ylabel('Spending Score (1-100)')
        plt.legend()
        plt.show()
        '''
        
    #skóre?
    def ucelovaFunkce(self):
        print(self.hodnotaUceloveFunkce)
```

```python
ks = kStredu(4)
ks.vlozData(X)
ks.inicializujStredy([0,3,5,2])
print(ks.stred)
print(ks.hodnotaUceloveFunkce)
ks.prirazeni()
ks.optimalizace()
print(ks.stred)
print(ks.hodnotaUceloveFunkce)
ks.prirazeni()
ks.optimalizace()
print(ks.stred)
print(ks.hodnotaUceloveFunkce)
ks.prirazeni()
ks.optimalizace()
print(ks.stred)
print(ks.hodnotaUceloveFunkce)
ks.prirazeni()
ks.optimalizace()
print(ks.stred)
print(ks.hodnotaUceloveFunkce)
print(ks.prislusnost)
ks.vykresliShlukovani()
```
