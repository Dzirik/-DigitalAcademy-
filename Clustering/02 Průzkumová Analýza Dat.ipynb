## Předzpracování dat

Na základě výše uvedené prvotní analýzy bylo zjištěno následující.

* Data obsahují outlier - jedná se o jedno pozorování a proto bude vyřazeno.
* Data mají špatné měřítko, šířka jse v cm, délka je mm. Rozhodnuto o převedení na cm.
* Přejmenování sluopců s ohledem na jednotky.
* Data obsahují jednu NaN hodnotu - bude vyřazena.

Vytvořte **nový** data frame **cistaData**, kde budou vyřešeny výše uvedené problémy.

> Pozn.: Outlier a NaN value jsou dva poslední řádky data framu.

```python
#vytvořím kopii bez posledních dvou pozorování - outlier a řádek obsahující NaN
cistaData=hrubaData[0:200].copy()

#převod z mm na cm
cistaData["Delka"]=cistaData["Delka"]/10

#přejmenování
cistaData.rename(columns={"Sirka":"Sirka [cm]","Delka":"Delka [cm]"},inplace=True)
```

```python
#zobrazení
cistaData.head()
```

```python
cistaData.tail()
```

```python
# NaN
print(cistaData.isnull().sum())
cistaData[cistaData["Sirka"].isnull()]
```

## Druhé vykreslení dat

Analogicky předchozí situaci vykreslete data.

```python
#vykreslení

#celé níže smazat ----------------------

#vytáhnu si obě datové sady - lepší opakovatelnost kódu.
x=cistaData["Sirka [cm]"]
y=cistaData["Delka [cm]"]

#základ obrázku
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])

#bodový obrázek
ax.scatter(
        x=x,
        y=y,
        c=barva[3],
        s=30,
        alpha=0.5,
        edgecolors=barva[0]
)

#pojmenování os
ax.set_xlabel(x.name)
ax.set_ylabel(y.name)
ax.set_title("Vizualizace čistých dat")

#nastavení os
plt.axis("equal")
#ax.set_xlim([])
#ax.set_ylim([])
```
